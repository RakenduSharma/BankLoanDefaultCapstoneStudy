#Spilt Data

```python
data = clean_df.sample(frac=0.001, random_state=12345)
data_unseen = clean_df.drop(data.index)
data.reset_index(drop=True, inplace=True)
data_unseen.reset_index(drop=True, inplace=True)
```

```python
print(data)
colsToDrop = ["member_id","funded_amnt_inv", "funded_amnt", "revol_bal", "revol_util","total_acc","installment","open_acc"]
data.drop(colsToDrop, axis=1, inplace=True)
data.reset_index(drop=True, inplace=True)
print(data)
```

```python
data.loc[(data.loan_status == 'Fully Paid'),'loan_status']=0
data.loc[(data.loan_status == 'Default'),'loan_status']=1
data['loan_status']= pd.to_numeric(data['loan_status'])
data.info()
observations = []
for i in range(len(data)):
 observations.append([str(data.values[i,j]) for j in range(11)])
```

#Apriori

```python
association_rules = apriori(observations, min_support=0.0045, min_confidence=0.2, min_lift=3, min_length=2)
association_results = list(association_rules)
```
